# Financial Strategy Response to 2022 Subsidy Reduction

This notebook reproduces the assumption-based analysis used in the case study.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
ROOT = Path('..').resolve()
kpi = pd.read_csv(ROOT / 'data' / 'kpi_summary.csv')
assumptions = pd.read_csv(ROOT / 'data' / 'assumptions_and_calculations.csv')
kpi.head()
```

## KPI Checks

```python
summary = {
    'profit_low_point_m': float(kpi['profit_m'].min()),
    'profit_latest_m': float(kpi['profit_m'].iloc[-1]),
    'cashflow_low_point_m': float(kpi['cashflow_m'].min()),
    'cashflow_latest_m': float(kpi['cashflow_m'].iloc[-1]),
    'avg_forecast_error_pct': float(kpi['forecast_error_pct'].mean()),
}
summary
```

## Intervention Lift (Assumption Model)

```python
assumptions[['quarter', 'assumed_mitigation_rate_pct', 'implied_intervention_lift_m']]
```

```python
plt.figure(figsize=(10,4.5))
x = np.arange(len(kpi))
plt.plot(kpi['quarter'], kpi['profit_m'], marker='o', label='Actual Profit ($M)')
plt.plot(assumptions['quarter'], assumptions['counterfactual_profit_without_intervention_m'], marker='o', label='Counterfactual: No Intervention ($M)')
plt.xticks(rotation=30)
plt.ylabel('USD Millions')
plt.title('Actual vs Counterfactual Profit (Assumption-Based)')
plt.legend()
plt.tight_layout()
plt.show()
```

## Interpretation

- The shock quarter starts in 2022-Q2.
- Mitigation rate improves quarter over quarter.
- By 2023-Q4, intervention lift is the highest in the period, consistent with the project narrative.
